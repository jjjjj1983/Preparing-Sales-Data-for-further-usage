```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# generate synthetic data for customer segmentation
np.random.seed(2024)
cluster_1 = np.random.normal([50, 5], [5, 1], size=(50, 2))
cluster_2 = np.random.normal([100, 20], [7, 2], size=(50, 2))
cluster_3 = np.random.normal([150, 40], [5, 1], size=(50, 2))
noise = np.random.uniform([0, 0], [200, 50], size=(20, 2))
data = np.vstack([cluster_1, cluster_2, cluster_3, noise])

# apply DBSCAN clustering
dbscan = DBSCAN(eps=10, min_samples=5)
labels = dbscan.fit_predict(data)

# plot the results
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
for label in unique_labels:
    if label == -1:  # noise points
        plt.scatter(data[labels == label, 0], data[labels == label, 1], color='red', label='Noise', alpha=0.6)
    else:
        plt.scatter(data[labels == label, 0], data[labels == label, 1], label=f'Cluster {label + 1}', alpha=0.7)

plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1 (e.g., Monthly Usage)")
plt.ylabel("Feature 2 (e.g., Service Cost)")
plt.legend()
plt.grid(True)
plt.show()
```
