#Both data series show non-normal destribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normaltest, shapiro

# Creating two data sets
np.random.seed(42)

salaries = [3000, 3200, 3500, 3700, 3100, 4000, 3300, 4500, 2900, 4200, 3400, 3100,
            2950, 3600, 3800, 5000, 3700, 5200, 4700, 5500, 4300, 6000, 3100, 2800,
            5000, 3200, 4100, 3900, 4500, 3000]
work_experience = [5.9, 4.9, 5.6, 7.3, 4.8, 4.8, 7.7, 6.6, 3.2, 6.3, 3.9, 3.9, 5.5,
                   3.9, 7.3, 6.7, 5.3, 2.9, 6.6, 5.1, 4.6, 4.6, 5.6, 8.5, 3.7, 4.4,
                   6.4, 4.5, 5.5, 3.6]

_, p_value_normaltest_normal = normaltest(salaries)
_, p_value_normaltest_non_normal = normaltest(work_experience)

# Shapiro-Wilk test
_, p_value_shapiro_normal = shapiro(salaries)
_, p_value_shapiro_non_normal = shapiro(work_experience)

# Test results
print(f"Normaltest for normal data p-value: {p_value_normaltest_normal}")
print(f"Normaltest for non-normal data p-value: {p_value_normaltest_non_normal}")

print(f"Shapiro-Wilk for normal data p-value: {p_value_shapiro_normal}")
print(f"Shapiro-Wilk for non-normal data p-value: {p_value_shapiro_non_normal}")

# Plotting histograms to visualize data distributions
plt.figure(figsize=(12, 5))

# Normal distribution
plt.subplot(1, 2, 1)
plt.hist(salaries, bins=30, color="skyblue", edgecolor="black")
plt.title("Salaries")

# Non-normal distribution
plt.subplot(1, 2, 2)
plt.hist(work_experience, bins=30, color="salmon", edgecolor="black")
plt.title("Wok Experience")

plt.tight_layout()
plt.show()
```
